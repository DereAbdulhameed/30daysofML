```python
def load_and_prepare_dataset(csv_paths, audio_dir):
    # Initialize a dict to hold the processed datasets
    processed_datasets = {}

    # Iterate over each dataset type (train, test) and process
    for ds_type, csv_path in csv_paths.items():
        # Load the TSV file into a Pandas DataFrame
        df = pd.read_csv(csv_path, delimiter='\t', usecols=['path', 'sentence'])

        # Convert the DataFrame to a Hugging Face Dataset
        dataset = Dataset.from_pandas(df)

        # Function to load and add the audio file to the dataset
        def add_audio_column(example):
            audio_path = audio_dir + '/' + example['path']  # Ensure the path is correct
            audio, _ = librosa.load(audio_path, sr=16000)
            # Convert the audio to a NumPy array
            audio = audio.astype(np.float32)
            example['audio'] = audio
            return example


        # Apply the function to add the audio column
        dataset = dataset.map(add_audio_column, remove_columns=['path'])

        # Rename 'sentence' column to 'transcription' for clarity
        dataset = dataset.map(lambda x: {'transcription': x['sentence'], 'audio': x['audio']}, remove_columns=['sentence'])

        # Store the processed dataset
        processed_datasets[ds_type] = dataset

    return processed_datasets

# Define the paths to your training and testing TSV files
csv_paths = {
    'train': '/content/train.tsv',
    'test': '/content/test.tsv'
}
audio_dir = '/content/drive/MyDrive/clips'

# Load and prepare the dataset
datasets = load_and_prepare_dataset(csv_paths, audio_dir)
```

```python
from torch.utils.data import Dataset

class YorubaSpeechDataset(Dataset):
    def __init__(self, dataset, processor):
        self.dataset = dataset
        self.processor = processor

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        audio, _ = librosa.load(item["path"], sr=16000)
        input_values = self.processor(audio, sampling_rate=16000, return_tensors="pt").input_values.squeeze()
        labels = self.processor.tokenizer(item["transcription"], padding="max_length", max_length=128, truncation=True).input_ids
        labels = [label if label != self.processor.tokenizer.pad_token_id else -100 for label in labels]
        return {"input_values": input_values, "labels": torch.tensor(labels)}
```

```python
from IPython.display import display, Javascript, HTML
from google.colab.output import eval_js
import base64
import io

def record_audio(duration=5):
    """
    Display a Javascript interface for recording audio in the notebook.
    Duration is the length of the recording in seconds.
    """
    display(Javascript('''
    async function recordAudio(duration) {
        // Create audio stream
        const stream = await navigator.mediaDevices.getUserMedia({audio: true});
        const options = {mimeType: 'audio/webm'};
        const mediaRecorder = new MediaRecorder(stream, options);
        let audioChunks = [];

        // Start recording
        mediaRecorder.start();

        // Push chunks of audio to array
        mediaRecorder.addEventListener('dataavailable', event => {
            audioChunks.push(event.data);
        });

        // Resolve promise with audio blob when recording stops
        const audioBlobPromise = new Promise(resolve => {
            mediaRecorder.addEventListener('stop', () => {
                const audioBlob = new Blob(audioChunks);
                resolve(audioBlob);
            });
        });

        // Stop recording after duration
        setTimeout(() => mediaRecorder.stop(), duration * 1000);

        // Return audio blob
        return await audioBlobPromise;
    }
    '''))

    # Record audio and get audio data as a base64 string
    audio_data = eval_js('''
    (async () => {
        const audioBlob = await recordAudio(%d);
        const reader = new FileReader();
        const audioBase64Promise = new Promise(resolve => {
            reader.onloadend = () => resolve(reader.result);
            reader.readAsDataURL(audioBlob);
        });
        return await audioBase64Promise;
    })()
    ''' % duration)

    # Convert base64 audio data to bytes
    audio_bytes = base64.b64decode(audio_data.split(',')[1])

    return audio_bytes
```

```python
import torch
from transformers import AutoModelForCTC, AutoProcessor

# Load your trained model and processor
model_name = "/content/yoruba_speech_model"
processor = AutoProcessor.from_pretrained(model_name)
model = AutoModelForCTC.from_pretrained(model_name)

def transcribe(audio_bytes):
    """
    Transcribe the provided audio bytes using the loaded model and processor.
    """
    # Load audio with librosa or any library that suits the model's requirement
    audio, sr = librosa.load(io.BytesIO(audio_bytes), sr=16000)

    # Preprocess the audio
    input_values = processor(audio, return_tensors="pt", sampling_rate=16000).input_values

    # Inference
    with torch.no_grad():
        logits = model(input_values).logits

    # Decode the logits to text
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = processor.batch_decode(predicted_ids)[0]

    return transcription
```

```python
# Record audio (adjust duration as needed)
audio_bytes = record_audio(duration=5)

# Transcribe the recorded audio
transcription = transcribe(audio_bytes)

# Display the transcription
print("Transcribed Text:", transcription)
```
